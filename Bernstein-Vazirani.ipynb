# Bernstein-Vazirani

```python
import numpy as np

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute
import learn_quantum as lq

```

```python
bit_string='100011'
oracle_size = len(bit_string)

oracle = QuantumRegister(oracle_size,'ora')
function = QuantumRegister(1, 'func')
cr_oracle = ClassicalRegister(oracle_size, 'out')

qc = QuantumCircuit(oracle, function, cr_oracle)

qc.h(oracle)

qc.x(function)
qc.h(function)
qc.x(function)

qc.barrier()

for i in range(oracle_size):
    if bit_string[i]=='1':
        qc.cx(oracle[i], function[0])

qc.barrier()
qc.h(oracle)

qc.measure(oracle, cr_oracle)

answers = lq.execute_simulated(qc, 1)
lq.print_reverse_results(answers, integer=False)
```
